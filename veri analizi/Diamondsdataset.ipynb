# Getting & Investigating the Data
In this section, we will be using Diamonds dataset which contains the information about diamonds.

Dataset has 10 columns which are:

 * **carat:** weight of the diamond (0.2--5.01)
 * **cut:** quality of the cut (Fair, Good, Very Good, Premium, Ideal)
 * **color:**diamond colour, from J (worst) to D (best)
 * **clarity:** a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
 * **depth:** total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
 * **table:** width of top of diamond relative to widest point (43--95)
 * **price:** price in US dollars ($326--$18,823)
 * **x:** length in mm (0--10.74)
 * **y:** width in mm (0--58.9)
 * **z:** depth in mm (0--31.8)

**For further information about this dataset, you can visit:** https://www.kaggle.com/shivam2503/diamonds

# Bar Graphs (Çubuk Grafikler)
Değerlerin nasıl dağıldığını görsel olarak araştırmak için kümelenmiş sayısal değerlerin histogramlarını çizmek ve ayrı bir öğe grubunun miktar dalgalanmasını göstermek için kullanılır.

**Horizontal Bar Plot (Yatay Çubuk Bar)**

```python
'''
the letter 'h' at the end of the method "barh()" stands for "horizontal"   "barh()" yönteminin sonundaki "h" harfi "yatay" anlamına gelir     '''
'''set_title() is a function of Matplotlib library                          set_title(), Matplotlib kitaplığının bir işlevidir
'''
df_diamonds["cut"].value_counts().plot.barh().set_title("Class Frequencies of the Cut Variable")    #set_title grafiğe kendimiz isim veriyoruz
```

**Vertical Bar Plot (Dikey Bar Plot)**

```python
'''
You can also customize the plot style
'''
sns.barplot(x="cut", y=df_diamonds["cut"].index,data=df_diamonds, color = 'purple').set_title("Cut Density")     # y=df_diamonds["cut"].index (y nin içine cut içindeki index değerlerini yaz)
```

**Grouped Bar Chart (Gruplandırılmış Çubuk Grafik)**
  * Amaç, çubuk grafiği oluşturan değişkenlerden birinin alt gruplarını göstermek olduğunda kullanılır.

* Başka bir değişkenin ton eşlemesi ile koşullu dağılımları çizer

* Renkler ve sıralamaları her zaman birbiriyle tutarlıdır.

```python
'''
You can indicate more than one categorical data on a plot. 
Be careful about the usage of inputs that indicates categorical data such as hue, size. You must use
categorical data column with them.
'''
sns.barplot(x="cut",
            y="price",
            hue="color",
            data=df_diamonds)
plt.title("Cut - Price - Color")
```

```python
sns.barplot(x="cut",
            y="price",
            hue="clarity",          # hue parametresi sol üstteki darity  renk kodunu gösterir
            data=df_diamonds)
plt.title("Cut - Price - Clarity")
```

**Scatter Plot (Dağılım grafiği)**


```python
cols = ["x", "y", "z"]
for col in cols:
  plt.figure(figsize=(10, 10))
  sns.scatterplot(x=col, y="price", data=df_diamonds)
  plt.show()
```

**Histogram**

 * Histogram, verilerin parametre olarak verilen kutu sayısına göre bölümlere ayrıldığı bir diyagramdır.

 * Seaborn kitaplığının histplot() işlevi, belirtilmediği sürece kutu sayısını otomatik olarak atar

```python
sns.histplot(df_diamonds["price"], 
             kde=True)  # kde = kernel density estimation  # bins = "auto"
```

```python
sns.histplot(df_diamonds["price"], bins=100, kde=True)  # bins is set to 100, not auto
```

```python
sns.histplot(df_diamonds["price"])  # the line is not visible, kde is set to False by default
```

```python
# displays only the line depicting the density 
sns.kdeplot(df_diamonds["price"])
```

```python
sns.histplot(df_diamonds["price"], 
             kde=False)
```

```python
sns.histplot(df_diamonds["price"], 
             bins=100, 
             kde=True)
```

**Joint Plot**
 * Bir ortak grafik, varsayılan olarak, verilen değişkenlerin 2 histogramından ve bunların korelasyonunu temsil eden 1 dağılım grafiğinden oluşur.

 * Parametreleri değiştirerek, dağılım grafiği yerine KDE grafiği de kullanılabilir.

```python
sns.jointplot(x="price", y=df_diamonds["carat"], data=df_diamonds)
```

**Catplot**

Kategorik Çizim, kategorik bir değişkenin öğelerine karşılık gelen sayısal bir değişkenin saçılımını görselleştirmek için kullanılır.

```python
sns.catplot(x="cut", y="price", data=df_diamonds)
plt.title("Cut - Price")
```

```python
(sns
.catplot(x="cut",
         y="price",
         hue="color",
         kind="point",
         data=df_diamonds))
plt.title('Cut - Price - Color')
```

***KDE Plot (Kernel Density Estimate) - KDE Grafiği (Çekirdek Yoğunluğu Tahmini)***

 * KDE Grafiği, sürekli bir değişkenin Olasılık Yoğunluğunu görselleştirmek için kullanılır

 * Tek bir grafikte birden fazla grafik de görüntülenebilir

```python
sns.kdeplot(df_diamonds["price"], shade=True)  # With shade=True, one can fill the space staying under the line
```

```python
# Investigate whiich value of a chosen categorical column affects how the distribution
(sns
 .FacetGrid(df_diamonds,
            hue="cut", 
            height=5,
            xlim = (0, 10000))
.map(sns.kdeplot, "price", shade=True)
.add_legend())
plt.title('Cut - Price')
```

```python
(sns
.FacetGrid(df_diamonds,
          hue="color",
          height=5,
          xlim = (0, 10000))
.map(sns.kdeplot,"price",shade=True)
.add_legend())
plt.title('Color - Price')
```

**Correlation Matrix (Korelasyon Matrisi)**

 * Bir korelasyon ısı haritası, iki sütun arasında bir 2B korelasyon matrisi (tablo) göstermek için renkli hücreler kullanır.

 * Korelasyon açısından, 1, iki sütun arasında güçlü bir pozitif korelasyonu temsil ederken, -1, bunlar arasında güçlü bir negatif korelasyonu temsil eder.

 * Korelasyon matrisinde verilen değerlere göre iki değişkenli analizin hangi sütunlarda gerçekleştirileceğini seçmek daha iyi bir yöntemdir.

```python
plt.title('Correlation Matrix')
sns.heatmap(df_diamonds.corr(), annot=True)  # heatmap() from sns library is used to plot the correlation matrix
```
